```python
import pandas as pd
import json

# Paso 2: Analizar el JSON en la columna "Value"
def parse_json(value):
    try:
        return json.loads(value)
    except:
        return {}

df['Value'] = df['Value'].apply(parse_json)

# Paso 3: Convertir el diccionario en un DataFrame
value_df = pd.json_normalize(df['Value'])

# Paso 4: Concatenar el nuevo DataFrame con el original y eliminar la columna JSON original
df = pd.concat([df, value_df], axis=1)

# Mostrar el DataFrame resultante
display(df)

```

```python
import pandas as pd

# Supongamos que tienes tu DataFrame llamado df

# Convertir la columna 'TS' a tipo datetime
df['TS'] = pd.to_datetime(df['TS'])

# Eliminar la información de la zona horaria
df['TS'] = df['TS'].dt.tz_localize(None)
```
