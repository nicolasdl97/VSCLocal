```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel(r"C:\Users\nicol\OneDrive\Documentos\VSCLocal\Data\Mari\LPS_binding.xlsx")
print(list(df.columns))
display(df.head())
```

```python
df_melt=df.melt(id_vars="uM",var_name="Peptide")
print(df_melt.head())
```

```python
df_Magainin=df_melt[df_melt["Peptide"].str.contains('Magainin', case=False, na=False)]
print(df_Magainin.shape)

df_CLS001=df_melt[df_melt["Peptide"].str.contains('CLS001', case=False, na=False)]
print(df_CLS001.shape)

df_Histatin=df_melt[df_melt["Peptide"].str.contains('Histatin', case=False, na=False)]
print(df_Histatin.shape)

df_Ctn_15_34=df_melt[df_melt["Peptide"].str.contains('Ctn 15-34', case=False, na=False)]
print(df_Ctn_15_34.shape)

df_Indolicidin=df_melt[df_melt["Peptide"].str.contains('Indolicidin', case=False, na=False)]
print(df_Indolicidin.shape)

df_Cecropin=df_melt[df_melt["Peptide"].str.contains('Cecropin', case=False, na=False)]
print(df_Cecropin.shape)

df_Melittin=df_melt[df_melt["Peptide"].str.contains('Melittin', case=False, na=False)]
print(df_Melittin.shape)

df_Bac8c=df_melt[df_melt["Peptide"].str.contains('Bac8c', case=False, na=False)]
print(df_Bac8c.shape)

df_means=df_melt[df_melt["Peptide"].str.contains('mean', case=False, na=False)]
print(df_Bac8c.shape)

dfs=[df_Magainin,df_CLS001,df_Histatin,df_Ctn_15_34,df_Indolicidin,df_Cecropin,df_Melittin,df_Bac8c,df_means]
```

```python
sns.set_style("whitegrid")
sns.set_palette("muted")
sns.set_context("notebook")
sns.relplot(data=df_Magainin,kind="line",y="uM",x="value",hue="Peptide",)
plt.show()
```

```python
sns.set_style("whitegrid")
sns.set_palette("muted")
sns.set_context("notebook")

for df in dfs:
    sns.relplot(data=df,kind="line",y="uM",x="value",hue="Peptide",)
    plt.show()
```
