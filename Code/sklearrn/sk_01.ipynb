```python
matriz=df.to_numpy()
```

```python
n= len(matriz)
sum_x= np.sum(matriz[:,0])
sum_y= np.sum(matriz[:,1])
suma_prod=np.sum(matriz[:,0]*matriz[:,1])
suma_cuadrado_x=np.sum(matriz[:,0]*matriz[:,0])

print("sum_x:",sum_x)
print("sum_y:",sum_y)
print("suma_prod:",suma_prod)
print("suma_cuadrado_x:",suma_cuadrado_x)
```

```python
b1=(n*suma_prod-sum_x*sum_y) / (n*suma_cuadrado_x-sum_x**2)
b0=(sum_y-b1*sum_x) / n
print("b1:",b1)
print("b0:",b0)
```

## scikit-learn

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from sklearn.metrics import root_mean_squared_error
clf=linear_model.LinearRegression(fit_intercept=True)
```

```python
n=len(x.values)
_x=df['Reducción de sólidos'].values.reshape(n,1)
_y=df['Reducción de la demanda de oxígeno'].values.reshape(n,1)
```

```python
clf.fit(_x,_y)
```

```python
print("Coefficients:", clf.coef_)
print("Intercepto:", clf.intercept_)
print("MSE: %f" % mean_squared_error(_y,clf.predict(_x)))
print("RMSE: %f" % root_mean_squared_error(_y,clf.predict(_x)))
```

```python
plt.plot(_x,_y)
plt.plot(_x,clf.predict(_x))
plt.title("Regresión Lineal Simple")
plt.xlabel("Reducción de solidos")
plt.ylabel("Reducción de demanda de oxígeno")
plt.legend(["y","predicciones"])
plt.grid()
plt.show()
```
